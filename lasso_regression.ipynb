```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('/content/sample_data/mydata.txt')
data
```

```python
df=data.copy()
df
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import Ridge
```

```python
l=50
ridge=Ridge(alpha=l)
ridge.fit(X,y)
```

```python
ridge_cof=ridge.coef_
ridge_cof
```

```python
ridge_inter=ridge.intercept_
ridge_inter
```

```python
X_test=[1,2,3,4]
y_pred=ridge.predict(pd.DataFrame(X_test))
y_pred
```

```python
plt.figure(figsize=(12,6),dpi=90)
plt.plot(X_test,y_pred,c='r', linewidth=2)
plt.scatter(X,y)
plt.xlim(xmin=0,xmax=7)
plt.ylim(ymin=0,ymax=9)
plt.text(X_test[-1],y_pred[-1],'y='+str('%.2f'%ridge_cof)+'*X+'+str('%.2f' %ridge_inter)
              +'For\u03BB or \u03b1='+str(l),fontsize=12)
plt.title('Ridge Regression')
plt.show()
```
